```python
import xarray as xr
from netCDF4 import Dataset
from xclim.indices import heating_degree_days
from xclim.indices import cooling_degree_days
```

```python
df=xr.open_dataset("ENSEMBLE_BC_tmean_Rcp45_2006_2099_dgC_India.nc")
```

```python
df["tmean"]=(df["tmean"]*1.8)+32
```

```python
#df = df.sortby('time')
```

```python
# Even we have converted to ferenhite but still we will be using degc attribute for package requirement
```

```python
df.tmean.attrs["units"]="degC"
df.tmean
```

```python
cdd = cooling_degree_days(df.tmean, 79.7) # Here we have added 79.7 F (26.5 degC) 
cdd.to_netcdf("ENSEMBLE_BC_CDD_Rcp45_2006_2099_dgF_India.nc")
```

```python
hdd = heating_degree_days(df.tmean,61.7)
hdd.to_netcdf("ENSEMBLE_BC_HDD_Rcp45_2006_2099_dgF_India.nc")
```

```python
def read_model(x):
    ncfile=x
    fh=Dataset(ncfile,mode='r')
    rainfall=fh.variables['tmean'][ : :]
    rainfall=rainfall.reshape(1*129,135)
    rainfall[rainfall<0]=np.nan
    return rainfall
```
